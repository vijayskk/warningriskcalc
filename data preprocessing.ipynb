The dataset contains two columns: "message" and "risk_level." The "message" column represents the terminal warning or error message, while the "risk_level" column indicates the risk level associated with the message. The risk levels are categorized as "low," "medium," or "high" to denote the severity of the warning or error.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data.csv")
df
```

```python
rl = []
for i in df['risk_level']:
    if i == 'low':
        rl.append(1)
    if i == 'medium':
        rl.append(2)
    if i == 'high':
        rl.append(3)
df['rl'] = rl
df
```

```python
mess = []
for i in df['message']:
    mess.append(i[:-1])
df['message'] = mess
df
```

```python
df.to_csv('processed_data.csv')
```
